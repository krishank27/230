# Pipeline Step 2: Importing well information data

```python
# As in Step 1, import the CSVs where the well information data is and combine into a single dataframe. 
# Ignore the 2007 file, the database it comes from is broken. 
import pandas as pd 
import glob
import matplotlib.pyplot as plt
%matplotlib inline

path ='/Users/krishankumar/Documents/csv' 
allFiles = glob.glob(path + "/*Wells.csv")

df = []

for f in allFiles:
    if '2007' in f:
        continue
    print(f)
    df.append(pd.read_csv(f , encoding = "ISO-8859-1", low_memory = False, dtype={'APINumber':str}))
```

```python
dfall = pd.concat(df, ignore_index=True)
```

```python
df[1].shape
```

```python
# We pull the relevant columns from the table. Wells are generally identified by their API number,
# but the wells are identified by PWT_ID in the production files. To match the correct API number to the corresponding
# data in the production data frame, we pull these two columns. 
temp = dfall[['APINumber','PWT__ID']].drop_duplicates()
```

```python
temp.info()
```

```python
temp.to_csv('wells.csv', index=False)
```
